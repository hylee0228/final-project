**Introduction**

*   Background

  <p>On October 22nd of 2023, the U.S., South Korea, and Japan held the first-ever trilateral military exercise near the Korean Peninsula. The aerial exercise, which included a nuclear-capable B-52 bomber, aimed to signal a response to continued military threats from North Korea. Although South Korea and Japan have been U.S. allies since the early 1950s separately, forming a trilateral alliance had been a tough agenda due to the history of the Japanese invasion and colonization of Korea from 1910 to 1945 which involved the exploitation of the economy and people, violence, harsh measures to eliminate cultural heritage, and most sensitive issue – sexual slavery. Often compared with Germany that apologized to countries for starting World War II, South Korea demanded the Government of Japan to apologize was never really met. Therefore, it has been a highly politicized issue as the general people’s sentiment remained cold when it comes to developing an open and cooperative relationship with the Japanese government.</p>
  <p>On the other hand, the U.S. historically sought US-ROK-Japan trilateral cooperation to respond to North Korean threats or to influence China’s behavior (Congressional Research Service, 2019) but a tense relationship with ROK-Japan prevented it. The Biden Administration has attempted to promote a trilateral military alliance since the beginning of the term as a key U.S. strategy toward East Asia politics and security. During the Biden Administration, South Korea had a change in government each of whom Biden had presidential meetings that resulted in a joint statement to lay out the principles of the U.S.-South Korea alliance. For instance, the 13th and former President of South Korea, Moon Jae-in met with President Joe Biden in May of 2021. The 14th and current President of South Korea, Yoon Suk-yeol, met with President Biden in August of 2023. The change in leadership in South Korea also changed the ruling party from liberal to conservative. Liberal parties have historically demonstrated anti-Japan sentiment whereas the conservative party was more favorable towards the Japanese government. Whereas an attempt to hold a presidential meeting among Prime Minister Suga, Biden, and Moon failed back in 2019, the first-ever stand-alone summit of Leaders from the three countries was held on August 13, 2023, at Camp David. And it is evident as demonstrated by recent trilateral military exercises that the current administration in South Korea has adopted policies to increase national defense cooperation with Japan.</p>

*   Research questions
<p>Reflecting this, the project aims to answer two research questions and is related to the corresponding hypothesis. First, reflecting upon recent events, was there a shift of tone over political alliance towards trilateral cooperation as South Korean party control changed from liberal to conservative party in 2022? The project will analyze statements of presidential joint meetings.</p>
<p>Then, the research will also expand to perform the same analysis of the publications from renowned think tanks on the topic of East Asian politics and international studies. As the government-level strategy gears towards positive comments on the trilateral alliance, think tanks would also publish their opinions that reflect it, acting to reinforce and support such a decision. Think tanks utilize expertise to advocate for specific policy positions and influence government decisions (Rich, 2005). Moreover, multiple case studies show how security and international affairs think tanks in emerging powers collaborate with policy-makers to meet current and anticipated foreign policy and security challenges (McGann, 2018). As the government promotes trilateral alliances, think tank reports reinforce and support government policy and vice versa.</p>
*   Hypothesis

  1.   As increasing media (news, articles) coverage over the trilateral meetings and activities shows, it implies that there was a shift in tone over trilateral cooperation among the U.S.-ROK-Japan that can be evident in the presidential meeting statements.
  2.   Major think tanks (i.e., Center for Strategic and International Studies) that cover the US-Korea security issues would show a similar trend to publish reports that contain the topic of trilateral cooperation over the same time span.

**Methods**
<p>The research aims to use various text analysis tools in Python to capture a shift in rhetoric over trilateral cooperation. Web scraping and text analysis will be used to perform this research.  It also contains remarks from presidential meetings which was the case after each of the U.S. and South Korean presidential meetings, including 13th President Moon Jae-in and 14th President Yoon Suk-Yeol.</p>

*   Using Python tools to analyze and compare the joint statements from the previous president to the current one, it is expected the frequency of  “Japan” would increase in recent statements issued after the meeting with the current president. Count features in nltk for key words will be used as well as dispersion plot that includes a time series with y-axis with words and dates covering two different administrations in South Korea spanning from May 2021 and August 2023 statements.
*   In order to identify key words from the two statements and gain insights into the main ideas, a wordcloud library will be created after cleaning the web-scrapped data including making the speech lowercase and lemmatizing to exclude stop words.
*   TF-IDF, which stands for Term Frequency-Inverse Document Frequency, will be used to identity important keywords within seven reports over South Korea in international relations and politics topic that were published by CSIS during 2021 to 2023. TF-IDF can be used to identify important keywords within a document. Words with higher TF-IDF scores are likely to be more representative of the content and can be considered as key terms.
*   Topic modeling using the Tomotopy library is also used over the seven reports to identify the key terms that contribute the most to the representation of topics in the reports and to rank documents based on their relevance to a topic group that contains “japan”. By sorting the document, can provide insights into whether the more recent publication contains topic words including “japan” than those published when the previous government was in power. It could imply that the think tank reports correspond to the shift in the government.

```python
def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)
#how to add verb "will" to lemmatizing --> Question
```

```python
import altair as alt
import numpy as np

# Terms in this list will get a red dot in the visualization
term_list = ['war', 'peace']

# adding a little randomness to break ties in term ranking
top_tfidf_plusRand = top_tfidf.copy()
top_tfidf_plusRand['tfidf'] = top_tfidf_plusRand['tfidf'] + np.random.rand(top_tfidf.shape[0])*0.0001

# base for all visualizations, with rank calculation
base = alt.Chart(top_tfidf_plusRand).encode(
    x = 'rank:O',
    y = 'document:N'
).transform_window(
    rank = "rank()",
    sort = [alt.SortField("tfidf", order="descending")],
    groupby = ["document"],
)

# heatmap specification
heatmap = base.mark_rect().encode(
    color = 'tfidf:Q'
)

# red circle over terms in above list
circle = base.mark_circle(size=100).encode(
    color = alt.condition(
        alt.FieldOneOfPredicate(field='term', oneOf=term_list),
        alt.value('red'),
        alt.value('#FFFFFF00')
    )
)

# text labels, white for darker heatmap colors
text = base.mark_text(baseline='middle').encode(
    text = 'term:N',
    color = alt.condition(alt.datum.tfidf >= 0.23, alt.value('white'), alt.value('black'))
)

# display the three superimposed visualizations
(heatmap + circle + text).properties(width = 600)
```

```python
training_data = []
original_texts = []
titles = []

for file in files:
    text = open(file, encoding='utf-8').read()
    text_tokens = nltk.word_tokenize(text)
    nltk_text = nltk.Text(text_tokens)
    text_lower = [t.lower() for t in nltk_text if t.isalpha()]
    text_stops = [t for t in text_lower if t not in stops]
    text_string = ' '.join(text_stops)
    training_data.append(text_string)
    original_texts.append(text)
    titles.append(Path(file).stem)
```

```python
len(training_data), len(original_texts), len(titles)
```

```python
# Number of topics to return
num_topics = 10
# Numer of topic words to print out
num_topic_words = 5

# Intialize the model
model = tp.LDAModel(k=num_topics)

# Add each document to the model, after removing white space (strip)
# and splitting it up into words (split)
for text in training_data:
    model.add_doc(text.strip().split())

# The log-likelihood function is typically used to
# derive the maximum likelihood estimator of the parameter
print("Topic Model Training...\n\n")
# Iterate over the data 10 times
iterations = 10
for i in range(0, 100, iterations):
    model.train(iterations)
    print(f'Iteration: {i}\tLog-likelihood: {model.ll_per_word}')


print("\nTopic Model Results:\n\n")
# Print out top 10 words for each topic
topics = []
topic_individual_words = []
for topic_number in range(0, num_topics):
    topic_words = ' '.join(word for word, prob in model.get_topic_words(topic_id=topic_number, top_n=num_topic_words))
    topics.append(topic_words)
    topic_individual_words.append(topic_words.split())
    print(f"✨Topic {topic_number}✨\n\n{topic_words}\n")
```

```python
topic_distributions = [list(doc.get_topic_dist()) for doc in model.docs]
```

```python
from IPython.display import Markdown, display
# IPython means interactive Python. It is an interactive command-line terminal for Python.
import re
# A RegEx, or Regular Expression, is a sequence of characters that forms a search pattern.
# RegEx can be used to check if a string contains the specified search pattern.

def make_md(string):
    display(Markdown(str(string)))

def get_top_docs(docs, topic_distributions, topic_index, n=5):
#     The zip() function takes iterables (can be zero or more), aggregates them in a tuple, and returns it.

    sorted_data = sorted([(_distribution[topic_index], _document)
                          for _distribution, _document
                          in zip(topic_distributions, docs)], reverse=True)

    topic_words = topics[topic_index]

    make_md(f"### ✨Topic {topic_index}✨\n\n{topic_words}\n\n---")

    for probability, doc in sorted_data[:n]:
        # Make topic words bolded
        for word in topic_words.split():
            if word in doc.lower():
                doc = re.sub(f"\\b{word}\\b", f"**{word}**", doc, re.IGNORECASE)

        make_md(f'✨  \n**Topic Probability**: {probability}  \n**Document**: {doc}\n\n')

    return
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='ticks', font_scale=1.2)
def plot_categories_by_topics_heatmap(labels,
                                      topic_distributions,
                                      topic_keys,
                                      output_path=None,
                                      target_labels=None,
                                      color_map = sns.cm.rocket_r,
                                      dim=None):

    # Combine the labels and distributions into a list of dictionaries.
#     The zip() function takes iterables (can be zero or more), aggregates them in a tuple, and returns it.
    dicts_to_plot = []
    for _label, _distribution in zip(labels, topic_distributions):
        if not target_labels or _label in target_labels:
            for _topic_index, _probability in enumerate(_distribution):
                dicts_to_plot.append({'Probability': float(_probability),
                                      'Category': _label,
                                      'Topic': 'Topic ' + str(_topic_index).zfill(2) + ': ' + ' '.join(topic_keys[_topic_index][:5])})

    # Create a dataframe, format it for the heatmap function, and normalize the columns.
    df_to_plot = pd.DataFrame(dicts_to_plot)
    df_wide = df_to_plot.pivot_table(index='Category',
                                     columns='Topic',
                                     values='Probability')
    df_norm_col=(df_wide-df_wide.mean())/df_wide.std()

    # Show the final plot.
    if dim:
        plt.figure(figsize=dim)
    sns.set(style='ticks', font_scale=1.2)
    ax = sns.heatmap(df_norm_col, cmap=color_map)
    ax.xaxis.tick_top()
    ax.xaxis.set_label_position('top')
    plt.xticks(rotation=30, ha='left')
    plt.tight_layout()
    if output_path:
        plt.savefig(output_path)
    plt.show()
```

```python
target_labels = titles
```

```python
plot_categories_by_topics_heatmap(titles,
                                  topic_distributions,
                                  topic_individual_words,
                                  target_labels=target_labels,
                                  color_map = 'Blues',
                                 dim=(12,9))
# For all possible color maps, see https://matplotlib.org/stable/tutorials/colors/colormaps.html#miscellaneous
```
